```
# Projeto caixa eletrônico.

# Você foi contratado para desenvolver um sistema simples de caixa eletrônico.

# O sistema deverá funcionar através de um menu

saldo = 1000.00
executando = True

while executando:
    print("\n=== CAIXA ELETRÔNICO ===")
    print("1. Consultar Saldo")
    print("2. Depositar")
    print("3. Sacar")
    print("4. Sair")

    opcao = input("Escolha uma opção (1-4): ")

    if opcao == '1':
        print(f"\nSeu saldo atual é R$ {saldo:.2f}")

    elif opcao == '2':
        valor = float(input("\nDigite o valor do depósito: R$ "))
        if valor > 0:
            saldo += valor
            print(f"Depósito realizado com sucesso! Novo saldo: R$ {saldo:.2f}")
        else:
            print("Valor inválido para depósito.")

    elif opcao == '3':
        valor = float(input("\nDigite o valor do saque: R$ "))
        if 0 < valor <= saldo:
            saldo -= valor
            print(f"Saque realizado com sucesso! Novo saldo: R$ {saldo:.2f}")
        elif valor > saldo:
            print("Saldo insuficiente.")
        else:
            print("Valor inválido para saque.")

    elif opcao == '4':
        print("\nObrigado por utilizar nosso Caixa Eletrônico. Volte sempre!")
        executando = False

    else:
        print("\nOpção inválida! Por favor, escolha um número de 1 a 4.")


```

```
# fazer um jogo de adivinhação

# o usuario ira dividir um numero de 0 até 10.

# regras

# o usuario tera 03 chances para adivinhar
# informar se o palpite do usuario é maior ou menor que o numero.
# imprimir mensagens de esta quente, morno ou frio.

import random

numero_secreto = random.randint(0, 10)
tentativas_restantes = 3

print("=========================================")
print("BEM-VINDO AO JOGO DE ADIVINHACAO!")
print("Tente adivinhar o numero entre 0 e 10.")
print("Você tem 3 chances. Boa sorte!")
print("=========================================\n")

while tentativas_restantes > 0:
    palpite = int(input("Digite seu palpite: "))

    if palpite == numero_secreto:
        print("MEUS MAIS SINCEROS PARABENS! Voce acertou! não fez mais do que sua obrigação O numero era", numero_secreto)
        break

    elif palpite < numero_secreto:
        print("Dica: O numero secreto e MAIOR.")

    elif palpite > numero_secreto:
        print("Dica: O numero secreto e MENOR.")

    distancia = abs(palpite - numero_secreto)

    if distancia == 1:
        print("Esta QUENTE!")
    elif distancia == 2:
        print("Esta MORNO!")
    elif distancia >= 3:
        print("Esta FRIO!")

    tentativas_restantes = tentativas_restantes - 1
    print("Tentativas restantes:", tentativas_restantes)
    print("-----------------------------------------")

if tentativas_restantes == 0 and palpite != numero_secreto:
    print("Fim de jogo! Voce perdeu mermão. O numero era", numero_secreto)

```
