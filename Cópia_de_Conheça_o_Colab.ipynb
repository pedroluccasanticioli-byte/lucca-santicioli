```
# 1 - Escrever um programa que o usuario digita um numero e o programa imprimi até o numero digitado.

numero = int(input("Digite um número inteiro: "))

print(f"Contando até {numero}:")

# O laço vai de 1 até numero + 1 (para incluir o próprio número)
for i in range(1, numero + 1):
    print(i)
```

```
# 2 - Escrever um programa que o usuario digita um numero e o programa imprimi o numero digitado até o zero (contagem regressiva).

numero = int(input("Digite um número para iniciar a contagem regressiva: "))

print(f"Iniciando contagem a partir de {numero}...")

# O loop 'for' vai do número digitado até -1 (para incluir o zero)
# O terceiro parâmetro '-1' define que o passo é decrescente
for i in range(numero, -1, -1):
    print(i)

print("Fim da contagem!")
```

```
# 3 - Escrever um programa que o usuario digita um numero e o programa imprimi os numeros pares.

limite = int(input("Digite um número inteiro: "))

print(f"Números pares de 0 até {limite}:")

# O laço vai de 0 até o número digitado (limite + 1 para incluir o próprio número)
for i in range(limite + 1):
    # Verifica se o número é divisível por 2 (resto é 0)
    if i % 2 == 0:
        print(i)
```

```
# 4 - Escrever um programa que o usuario digita um numero e o programa imprimi os numeros impares.

limite = int(input("Digite um número inteiro: "))

print(f"Números ímpares de 1 até {limite}:")

# O loop vai de 1 até o limite + 1
for i in range(1, limite + 1):
    # Verifica se o número é ímpar (resto da divisão por 2 é diferente de 0)
    if i % 2 != 0:
        print(i)
```

```
# 5 - Escrever um programa que o usuario digita um numero e programa soma os numeros até o numero digitado.

numero = int(input("Digite um número inteiro positivo: "))

# Inicializa a variável de soma
soma = 0

# Calcula a soma dos números de 1 até o número digitado
for i in range(1, numero + 1):
    soma += i

# Exibe o resultado
print(f"A soma de 1 até {numero} é: {soma}")
```

```
# 6 - Escrever um programa que o usuario digita uma palavra e o programa conta as letras da palavra.

palavra = input("Digite uma palavra: ")

# 2. Conta o número de caracteres usando a função len()
quantidade_letras = len(palavra)

# 3. Exibe o resultado
print(f"A palavra '{palavra}' tem {quantidade_letras} letras.")
```

```
# 7 - Escrever um programa que o usuario digita uma palavra e o programa conta as vogais dapalavra.

palavra = input("Digite uma palavra: ")

# Define quais caracteres são vogais (minúsculas e maiúsculas)
vogais = "aeiouAEIOU"
contador = 0

# Percorre cada letra da palavra
for letra in palavra:
    # Verifica se a letra está na string de vogais
    if letra in vogais:
        contador += 1

# Exibe o resultado
print(f"A palavra '{palavra}' possui {contador} vogais.")
```

```
# 8 - Escrever um programa que o usuario digita um numero e o programa imprimi a palavra invertida.

palavra = input("Digite uma palavra: ")

# Define quais caracteres são vogais (minúsculas e maiúsculas)
vogais = "aeiouAEIOU"
contador = 0

# Percorre cada letra da palavra
for letra in palavra:
    # Verifica se a letra está na string de vogais
    if letra in vogais:
        contador += 1

# Exibe o resultado
print(f"A palavra '{palavra}' possui {contador} vogais.")
```

```
# 9 - Escrever um programa que o usuario digita 04 numeros e o programa imprime o maior.

num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))
num3 = float(input("Digite o terceiro número: "))
num4 = float(input("Digite o quarto número: "))

# Encontrando o maior número usando a função max()
maior = max(num1, num2, num3, num4)

# Imprimindo o resultado
print(f"O maior número digitado foi: {maior}")
```
