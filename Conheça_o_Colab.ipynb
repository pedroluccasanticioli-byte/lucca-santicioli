```
# Solicite ao usuário um número inteiro positivo.

# Utilizando um laço de repetição, calcule e exiba a soma de
# todos os números de 1 até o número informado.

try:
    # Entrada de dados
    n = int(input("Digite um número inteiro positivo: "))

    # Validação de entrada
    if n <= 0:
        print("Erro: o número deve ser maior que zero.")
    else:
        soma = 0
        # Laço de repetição para somar de 1 até n
        for i in range(1, n + 1):
            soma += i

        # Exibe o resultado
        print(f"A soma de 1 até {n} é: {soma}")

except ValueError:
    print("Entrada inválida! Digite apenas números inteiros.")
```

```
# Solicite ao usuário um número inteiro e exiba sua tabuada
# de 1 até 10.

# 1 - Escrever um programa que o usuario digita um numero e o programa imprimi até o numero digitado.

numero = int(input("Digite um número inteiro: "))

print(f"Contando até {numero}:")

# O laço vai de 1 até numero + 1 (para incluir o próprio número)
for i in range(1, numero + 1):
    print(i)
```

```
# Solicite ao usuário um número inteiro positivo e calcule
# seu fatorial utilizando um laço de repetição.

numero = int (input("digite um numero:"))

for i in range (1, 11):
  print (f"{numero} x {i} = {numero * i}")
```

```
# Solicite ao usuário o valor do saque e informe a menor
# quantidade possível de notas para realizar o pagamento.

saldo = 1000.00
executando = True

while executando:
    print("\n=== CAIXA ELETRÔNICO ===")
    print("1. Consultar Saldo")
    print("2. Depositar")
    print("3. Sacar")
    print("4. Sair")

    opcao = input("Escolha uma opção (1-4): ")

    if opcao == '1':
        print(f"\nSeu saldo atual é R$ {saldo:.2f}")

    elif opcao == '2':
        valor = float(input("\nDigite o valor do depósito: R$ "))
        if valor > 0:
            saldo += valor
            print(f"Depósito realizado com sucesso! Novo saldo: R$ {saldo:.2f}")
        else:
            print("Valor inválido para depósito.")

    elif opcao == '3':
        valor = float(input("\nDigite o valor do saque: R$ "))
        if 0 < valor <= saldo:
            saldo -= valor
            print(f"Saque realizado com sucesso! Novo saldo: R$ {saldo:.2f}")
        elif valor > saldo:
            print("Saldo insuficiente.")
        else:
            print("Valor inválido para saque.")

    elif opcao == '4':
        print("\nObrigado por utilizar nosso Caixa Eletrônico. Volte sempre!")
        executando = False

    else:
        print("\nOpção inválida! Por favor, escolha um número de 1 a 4.")
```
